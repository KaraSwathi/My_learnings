```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["figure.figsize"]=(12.0,9.0)
# create data
x = np.arange(4) # Changed x to have 4 elements to match y1, y2, y3, y4, and y5
y1 = [0.78, 0.77, 0.77, 0.76]
y2 = [0.79, 0.78, 0.78, 0.78]
y3 = [0.79, 0.76, 0.77, 0.74]
y4 = [0.91, 0.89, 0.89, 0.90]
#y5 = [0.76, 0.65, 0.68, 0.66]
width = 0.2
group_spacing = 0.38
# plot data in grouped manner of bar type
plt.bar(x-0.45*group_spacing, y1, width, color='khaki')        #VGG16
plt.bar(x, y2, width, color='lightblue')                      #VGG19
plt.bar(x+0.19, y3, width, color='lightpink')                  #ResNet50
plt.bar(x+0.36, y4, width, color='plum')                       #CNN
#plt.bar(x+0.56, y5, width, color='turquoise')                  #TCNN
plt.xticks(x, ["Precision", "recall" , "f1 score","accuracy" ]) # Changed xtick labels to match the 4 data points
plt.xlabel("Network Architecture")
plt.ylabel("Metrics")
plt.legend(['VGG16','VGG19' ,'ResNet50' ,'CNN', 'TCNN'])
plt.ylim(0.00, 1.00)
plt.show()
plt.savefig('graph')
```
