```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('drug(1).csv')
df.head(10)
```

```python
df.shape
```

```python
X=df[['Age' ,	'Sex'	,'BP'	,'Cholesterol','Na_to_K']].values
X[0:5]
```

```python
from sklearn import preprocessing
le_sex = preprocessing.LabelEncoder()
le_sex.fit(['F','M'])
X[:,1] = le_sex.transform(X[:,1])

le_BP = preprocessing.LabelEncoder()
le_BP.fit([ 'LOW', 'NORMAL', 'HIGH'])
X[:,2] = le_BP.transform(X[:,2])

le_cholestral = preprocessing.LabelEncoder()
le_cholestral.fit([ 'NORMAL', 'HIGH'])
X[:,3] = le_cholestral.transform(X[:,3])

X[0:5]
```

```python
y = df["Drug"]
y[0:5]
```

```python
from sklearn.model_selection import train_test_split
X_trainset, X_testset, y_trainset, y_testset = train_test_split(X, y, test_size=0.3, random_state=3)
print(X_trainset.shape)
print(y_trainset.shape)
```

```python
print(X_testset.shape)
print(y_testset.shape)
```

```python
drugTree = DecisionTreeClassifier(criterion="entropy", max_depth = 4)
drugTree
```

```python
drugTree.fit(X_trainset,y_trainset)
```

```python
predtree = drugTree.predict(X_testset)
```

```python
print(predtree[0:5])
print(y_testset[0:5])
```

```python
from sklearn import metrics
import matplotlib.pyplot as plt
print("DecisionTrees's Accuracy: ", metrics.accuracy_score(y_testset, predtree))
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn import tree
result=confusion_matrix(y_testset, predtree)
print("Confusion Matrix:")
print(result)
```

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv('drug(1).csv')
df.head(10)
```

```python
df.shape
```

```python
X=df[['Age' ,	'Sex'	,'BP'	,'Cholesterol','Na_to_K']].values
X[0:5]
```

```python
d = pd.get_dummies(df['Sex'])
d.head()
```

```python
from sklearn import preprocessing
le_sex = preprocessing.LabelEncoder()
le_sex.fit(['F','M'])
X[:,1] = le_sex.transform(X[:,1])

le_BP = preprocessing.LabelEncoder()
le_BP.fit([ 'LOW', 'NORMAL', 'HIGH'])
X[:,2] = le_BP.transform(X[:,2])

le_cholestral = preprocessing.LabelEncoder()
le_cholestral.fit([ 'NORMAL', 'HIGH'])
X[:,3] = le_cholestral.transform(X[:,3])

X[0:5]
```

```python
y = df["Drug"]
y[0:5]
```

```python
from sklearn.model_selection import train_test_split
X_trainset, X_testset, y_trainset, y_testset = train_test_split(X, y, test_size=0.3, random_state=3)
print(X_trainset.shape)
print(y_trainset.shape)
```

```python
print(X_testset.shape)
print(y_testset.shape)
```

```python
classifier = RandomForestClassifier(n_estimators = 200,criterion='entropy',min_samples_leaf=5,random_state=0)
classifier.fit(X_trainset, y_trainset)
```

```python
y_pred1 = classifier.predict(X_testset)
from sklearn.metrics import  confusion_matrix
result = confusion_matrix(y_testset, y_pred1)
print("Confusion Matrix:")
print(result)
```
